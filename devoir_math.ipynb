```python
import matplotlib.pyplot as plt
import math
import numpy as np
from random import random
```

# Dessin des courbes f(x) et f'(x)

```python
def f(x):
    return np.exp(-x/10) * np.sin(x)
```

```python
def function (f,a,b,number_of_points):
    
    x_values = numpy.linspace(a,b,number_of_points)
    y_values = f(x_values)
    df_dx = np.gradient(y_values, x_values)
    
    plt.figure(figsize = (15,6))
    plt.plot(x_values, y_values, label='ma fonction')
    plt.plot(x_values, df_dx, label='la dérivée de ma fonction')
    plt.xlabel('les valeurs de x')
    plt.ylabel('les valeurs de y')
    plt.legend()
    plt.show()
    
    
    return x_values, y_values, df_dx
    
```

```python
a, b = 0, 10
number_of_points = 10000
x_values, y_values, df_dx = function(f,a,b,number_of_points)
```

# Le slicing et le calcule de la moyenne et l'équartype 

```python
def get_information_for_function (x_values,y_values,start_slice,end_slice,number_of_points):
   
    new_values = y_values[(x_values>=start_slice) & (x_values<=end_slice)]
    
    print(new_values)
    print('La moyenne est',np.mean(new_values))
    print('L équartype est',np.std(new_values))
    
    return new_values
```

# Méthode de Monte Carlo
o(X;Y) point aleatoire sur la courbe, Y<=f(X) signifie que o est inculu dans la surface 
en dessous de la courbe, p=Ai/At 
p=100000/10000=10 (100000 est la probabilité des points en dessous de la courbe)
At(la surface totale) = 10*0.875 = 8.75  

```python
def f(x):
    return np.exp(-x/10) * np.sin(x)

a, b = 0, 10
number_of_points = 10000
points_included = 0

max(f(a),f(b)) == 0.875
b-a == 10

for i in range(1, number_of_points + 1):
    X = random() * (b-a)
    Y = random() * max(f(a),f(b))
    if Y <= f(X):
        points_included += 1

estimated_area = 8.75 * (points_included / number_of_points)
print(estimated_area)
```
